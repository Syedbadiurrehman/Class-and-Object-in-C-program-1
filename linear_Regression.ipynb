```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error,root_mean_squared_error
```

```python
df = pd.read_csv('/content/Salary_dataset.csv')
print(df)
```

```python
sns.lmplot(x='YearsExperience', y='Salary', data=df)
```

```python
model = linear_model.LinearRegression()
model.fit(df[['YearsExperience']], df['Salary'])
```

```python
model.score(df[['YearsExperience']], df['Salary'])
```

```python
model.predict([[8.8]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y=model.coef_*8.8+model.intercept_
y
```

```python
mse = mean_squared_error(df['Salary'], model.predict(df[['YearsExperience']]))
print(mse)
```

```python
mae = mean_absolute_error(df['Salary'], model.predict(df[['YearsExperience']]))
print(mae)
```

```python
rmse = np.sqrt(mean_squared_error(df['Salary'], model.predict(df[['YearsExperience']])))
print(rmse)
```
